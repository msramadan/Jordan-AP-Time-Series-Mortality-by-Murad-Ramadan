```python
data['تاريخ الولادة'] = data['تاريخ الولادة'].astype('datetime64[ns]')
data['تاريخ الوفاة'] = data['تاريخ الوفاة'].astype('datetime64[ns]')
```

```python
data['Age'] = np.round((data['تاريخ الوفاة'] - data['تاريخ الولادة']).dt.days / 365.25, 2)
```

```python
data['Total >30'] = 0
data.loc[(data['Age'] >= 30), 'Total >30'] = 1
data['Total_Male >30'] = 0
data.loc[(data['Age'] >= 30) & (data['الجنس'] == 'ذكر'), 'Total_Male >30'] = 1
data['Total_Female >30'] = 0
data.loc[(data['Age'] >= 30) & (data['الجنس'] == 'انثى'), 'Total_Female >30'] = 1
data['>65'] = 0
data.loc[(data['Age'] >= 65), '>65'] = 1
data['>65_Male'] = 0
data.loc[(data['Age'] >= 65) & (data['الجنس'] == 'ذكر'), '>65_Male'] = 1
data['>65_Female'] = 0
data.loc[(data['Age'] >= 65) & (data['الجنس'] == 'انثى'), '>65_Female'] = 1
```

```python
data['Premature'] = 0
```

```python
data.loc[(data['Age'] < 77.81) & (data['Age'] > 30), 'Premature'] = 1
```

```python
data['Premature_Male'] = 0
data.loc[(data['Premature'] == 1) & (data['الجنس'] == 'ذكر'), 'Premature_Male'] = 1
```

```python
data['Premature_Female'] = 0
data.loc[(data['Premature'] == 1) & (data['الجنس'] == 'انثى'), 'Premature_Female'] = 1
```

```python
data['YLL'] = 0
data.loc[data['Premature'] == 1, 'YLL'] = 77.81 - data['Age']
data['YLL_Male'] = 0
data.loc[(data['Premature'] == 1) & (data['الجنس'] == 'ذكر'), 'YLL_Male'] = 77.81 - data['Age']
data['YLL_Female'] = 0
data.loc[(data['Premature'] == 1) & (data['الجنس'] == 'انثى'), 'YLL_Female'] = 77.81 - data['Age']
```

```python
data['Total'] = 1
data['Total_Male'] = 0
data.loc[data['الجنس'] == 'ذكر', 'Total_Male'] = 1
data['Total_Female'] = 0
data.loc[data['الجنس'] == 'انثى', 'Total_Female'] = 1
```
