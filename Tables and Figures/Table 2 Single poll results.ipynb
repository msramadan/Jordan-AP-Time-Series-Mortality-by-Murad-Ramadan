```python
import numpy as np
import pandas as pd
import os
```

```python
folder = './Total/'
files = sorted(os.listdir(folder))
```

```python
tot = [pd.read_excel(folder + file) for file in files]
```

```python
((tot[0]['Lag']).values)
```

```python
lag = ['0', '1', '2', '3', '4', '5', '6', '01', '25', '05']
```

```python
for i in range(len(tot)):
    tot[i]['Lag'] = lag
    tot[i].set_index('Lag', inplace=True)   
```

```python
tot[0]
```

```python
df = pd.DataFrame(index=lag)
for i in range(len(tot)):
    tot[i].fillna(' ', inplace=True)
    df[files[i]] = tot[i]['IR10'].round(2).astype(str) + ' (' + tot[i]['inf10'].round(2).astype(str) + '-' + tot[i]['sup10'].round(2).astype(str) + ')' + tot[i]['significance']
```
