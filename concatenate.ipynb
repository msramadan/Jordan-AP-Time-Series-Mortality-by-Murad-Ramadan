```python
import numpy as np
import pandas as pd
```

```python
pub_holi = pd.read_excel('jordan_public_holidays_with_islamic_dates_expanded.xlsx').set_index('Date')
```

```python
pub_holi['Holiday'] = 1
```

```python
amm_ap = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\AP data\Amman.xlsx").set_index('Date')
irb_ap = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\AP data\Irbid.xlsx").set_index('Date')
zar_ap = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\AP data\Zarqaa.xlsx").set_index('Date')
```

```python
amm_met = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\NC data\Excel\Amman.xlsx").set_index('time')
irb_met = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\NC data\Excel\Irbid.xlsx").set_index('time')
zar_met = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\NC data\Excel\Zarqa.xlsx").set_index('time')
```

```python
amm_mort = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\New mort\districts\Amman.xlsx").set_index('Date')
irb_mort = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\New mort\districts\Irbid.xlsx").set_index('Date')
zar_mort = pd.read_excel(r"C:\Users\user\OneDrive\Desktop\new ap mort\New mort\districts\Zarqa.xlsx").set_index('Date')
```

Amman

```python
pub_holi = pub_holi.reset_index().drop_duplicates(subset=['Date'], keep='first').set_index('Date')
```

```python
amm = pd.concat([amm_ap, amm_met, amm_mort, pub_holi], axis=1).dropna(subset=['t2m'])
```

```python
amm_mort
```

```python
amm.columns
```

```python
['Premature', 'Premature_Male',
       'Premature_Female', 'YLL', 'YLL_Male', 'YLL_Female', 'Total',
       'Total_Male', 'Total_Female', 'PM10', 'NO2', 'SO2', 't2m', 'sp', 'd2m', 'RH', 'Holiday']
```

```python
amm['Holiday'].fillna(0, inplace=True)
amm['Holiday'] = amm['Holiday'].astype(int)
amm['Date'] = amm.index
amm.reset_index(drop=True, inplace=True)
amm = amm[['Date', 'Total >30',
       'Total_Male >30', 'Total_Female >30', '>65', '>65_Male', '>65_Female',
       'Premature', 'Premature_Male', 'Premature_Female', 'YLL', 'YLL_Male',
       'YLL_Female', 'Total', 'Total_Male', 'Total_Female', 'Holiday', 'PM10', 'NO2', 'SO2', 't2m', 'sp', 'd2m', 'RH', 'Holiday']]
amm = amm[(amm['Date'] < '01-01-2020') & (amm['Date'] >= '01-01-2015')]
amm.to_excel('Amman.xlsx', index=False)
```

Irbid
